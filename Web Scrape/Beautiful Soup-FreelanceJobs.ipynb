 Perform simple web scraping on https://www.freelancer.in/ to gather freelancers Job,Description,Requirements and Bid Price data for analysis.

Using BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.freelancer.in/jobs/"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    jobs = soup.find_all('div', class_='JobSearchCard-item')

    job_list = []
    for job in jobs:
        try:
            job_title = job.find('a', class_='JobSearchCard-primary-heading-link').text.strip()
            description = job.find('p', class_='JobSearchCard-primary-description').text.strip()
            requirements = job.find('div', class_='JobSearchCard-primary-tags').text.strip()
            bid_price = job.find('div', class_='JobSearchCard-secondary-price').text.strip()
        except AttributeError:
            continue

        job_list.append({
            'Job Title': job_title,
            'Description': description,
            'Requirements': requirements,
            'Bid Price': bid_price
        })

    df = pd.DataFrame(job_list)
    df.to_csv('freelancer_jobs.csv', index=False)
    print("Data saved to freelancer_jobs.csv")
else:
    print("Failed to retrieve the webpage.")
```
